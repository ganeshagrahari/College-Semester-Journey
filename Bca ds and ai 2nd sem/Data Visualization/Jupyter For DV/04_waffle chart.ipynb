```python
import pandas as pd
import matplotlib.pyplot as plt 
from pywaffle import waffle
```

```python
the_data={
    'cricket':['sachin','sewag','sourav','gambir','rohit'],
    'centuries':[10,8,7,6,3]

}
df=pd.DataFrame(the_data)
fig=plt.figure(
    FigerClass=waffle,
    rows=5,
    values=df.centuries,
    label=list(df.cricket)
)
plt.show()
```
